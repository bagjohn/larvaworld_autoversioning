# Replaying experiments



Initialize the larvaworld registry. This loads some components from disc and builds the rest on the fly.

```python
import numpy as np

from larvaworld.lib import aux, reg, process
from larvaworld.lib.sim.dataset_replay import ReplayRun
reg.VERBOSE=1

```

```python
refIDs=reg.conf.Ref.confIDs
refID = refIDs[1]
# refID = 'Rehydration/AttP2.Deprived'
# refID = 'exploration.dish'
#refID = reg.default_refID
# refID = 'naive_locomotion.20controls'
# refID = 'exploration.150controls'
d=reg.loadRef(refID)
```

```python
def getConf(m, refID, d):
    replay_kws = {
    '2segs': {
            'draw_Nsegs': 2
        },
    'all_segs': {
            'draw_Nsegs': d.config.Npoints-1
        },
    'normal': {
        'time_range': (10,70)
        },
    'dispersal': {
        'transposition': 'origin',
        'time_range': (30,130)
        },
    'fixed_point': {
        # 'id':f'{refID}_replay_solo_fixed_point',
        'agent_ids': [1],
        'close_view': True,
        'fix_point': 6,
        },
    'fixed_segment': {
        # 'id':f'{refID}_replay_solo_fixed_point',
        'agent_ids': [1],
        'close_view': True,
        'fix_point': 6,
        'fix_segment': 'rear',
        },
    'fixed_overlap': {
        # 'id':f'{refID}_replay_solo_fixed_point',
        'agent_ids': [1],
        'close_view': True,
        'fix_point': 6,
        'fix_segment': 'rear',
        'overlap_mode': True,
        },
    }
    return reg.gen.Replay(**aux.AttrDict({
            'refID': refID,
            #'refDataset' : d,
            **replay_kws[m]
        })).nestedConf

```

```python
d.load()
d.enrich(proc_keys=['angular', 'spatial'])
s,e,c=d.data
s.columns
```

```python
# mode = '2segs'
# mode = 'fixed_overlap'
#mode = 'all_segs'
mode = 'normal'




rep = ReplayRun(parameters=getConf(m=mode, refID=refID,d=d),dataset=d, id=f'{refID}_replay_{mode}', dir= f'./media/{mode}')
output=rep.run()
```
